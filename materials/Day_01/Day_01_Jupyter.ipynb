## 3. Notebook as multimedia interface
### 3.1 Display image

```python
import IPython.display as disp
fhi_pc_logo_url = 'https://storage.googleapis.com/thehundreds/media/2017/06/convergence-1.jpg'
pollock = disp.Image(url=fhi_pc_logo_url)
disp.display(pollock)
```

```python
type(pollock)
```

```python
pollock.__dict__
```

```python
from PIL import Image
import urllib.request
import io
import numpy as np

with urllib.request.urlopen(fhi_pc_logo_url) as url:
    imgfile = io.BytesIO(url.read())

img_pollock = Image.open(imgfile)
img_pollock.__dict__
```

```python
pollock_data = np.asarray(img_pollock)
pollock_data.shape
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
f, axs = plt.subplots(2, 2, figsize=(16, 10))
axs[0,0].imshow(pollock_data)
axs[0,0].axis('off')
axs[0,1].imshow(pollock_data[...,0], cmap='Reds')
axs[0,1].axis('off')
axs[1,0].imshow(pollock_data[...,1], cmap='Greens')
axs[1,0].axis('off')
axs[1,1].imshow(pollock_data[...,2], cmap='Blues')
axs[1,1].axis('off')

plt.subplots_adjust(wspace=0.05, hspace=0.05)
```
